    <h4 style="color: Darkgreen;">
        <strong>
            17. What does the parameter inplace mean?
        </strong>
</div>

The inplace parameter in Pandas is used to specify whether an operation should modify the existing DataFrame (or Series) directly, or whether it should return a new object with the changes applied.


<div style="border: 2px Darkred; padding: 15px; background-color: #f9f9f9;">
    <h5 style="color: Black;">
        <strong>
            Example 1:
</div>

```python
# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Drop a column 'B' and modify the original DataFrame in place
df.drop('B', axis=1, inplace=True)

# Now df no longer has column 'B'
print(df)
```


<div style="border: 2px Darkred; padding: 15px; background-color: #f9f9f9;">
    <h5 style="color: Black;">
        <strong>
            Example 2:
</div>

```python
# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Drop column 'B' and return a new DataFrame (df is unchanged)
new_df = df.drop('B', axis=1)

# original df remains unchanged, new_df has column 'B' removed
print(df)
print(new_df)
```


<div style="border: 2px Darkred; padding: 15px; background-color: #f9f9f9;">
    <h5 style="color: Black;">
        <strong>
             Summary:
</div>

* inplace=False is the default value, so if you don't specify it, the operation will return a new copy of the DataFrame.
* inplace=True means the operation will be performed on the original DataFrame and modify it in place.

    <h4 style="color: Darkgreen;">
        <strong>
            18. Can a column in a table serve as both an index and a regular column?
        </strong>
</div>

Yes, a column in a table (DataFrame) in Pandas can serve as both an index and a regular column, but not simultaneously in the exact same form. <br> Here’s how this works:

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           1. Column as an Index:
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * When you set a column as the index of a DataFrame using set_index(), that column is used to uniquely identify the rows of the DataFrame.<br> The column is removed from the columns part of the DataFrame and becomes part of the index.

</div>

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           2. Column as a Regular Column:
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * A regular column is a field in the DataFrame that holds data and is part of the DataFrame structure.
</div>

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           3. Using a Column as Both:
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * To have a column serve as both an index and a regular column, you can set that column as the index temporarily and then add it back as a regular column using reset_index(). <br>This will "duplicate" the column, where the original data becomes part of the index and the same data will also appear as a regular column in the DataFrame.
</div>


<div style="border: 2px Darkred; padding: 15px; background-color: #f9f9f9;">
    <h5 style="color: Black;">
        <strong>
            Example:
</div>

```python
# Create a sample DataFrame
data = {
    'A': [1, 2, 3],
    'B': ['X', 'Y', 'Z'],
    'C': [10, 20, 30]
}
df = pd.DataFrame(data);

# Set column 'B' as the index
df.set_index('B', inplace=True);

# Now 'B' is the index, and it's no longer a regular column.
print("DataFrame with 'B' as index:");
print(df);

# Reset the index, which will bring 'B' back as a regular column
df_reset = df.reset_index();

# Now 'B' is both an index and a regular column
print("\nDataFrame with 'B' as both index and regular column:");
print(df_reset);
```

    <h4 style="color: Darkgreen;">
        <strong>
            19. What do axis 0 and axis 1 mean in a dataframe (df)?
        </strong>
</div>

In a Pandas DataFrame, axis is a parameter used to specify the direction along which a particular operation is to be applied. <br>The values 0 and 1 represent two different axes:

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           1. axis=0 (Rows, Vertical Axis):
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * axis=0 refers to operations along the rows of the DataFrame, i.e., it works vertically.<br>
        * Most commonly, this axis is used when you want to apply a function to each column (since columns are the units of data within rows).<br>
        * Operations like summing, averaging, or dropping rows are often done along axis=0.
</div>

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           1. Example Operations for axis=0:
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * Dropping rows: If you want to remove a row (or rows), you use axis=0.<br>
        * Applying functions across columns: When performing a sum or mean across columns, you use axis=0 to apply the operation across rows for each column
</div>

<div style="border: 2px Darkred; padding: 15px; background-color: #f9f9f9;">
    <h5 style="color: Black;">
        <strong>
            Example:
</div>

```python
df.sum(axis=0);
```

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           2. axis=1 (Columns, Horizontal Axis):
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * axis=1 refers to operations along the columns of the DataFrame, i.e., it works horizontally.<br>
        * When you use axis=1, you are typically applying a function across the rows for each column.<br>
        * Operations like summing, averaging, or dropping columns are typically done along axis=1
</div>

<div style="border: 2px solid Darkred; padding: 15px; background-color: #f9f9f9;">
    <h4 style="color: Darkred;">
        <strong>
           1. Example Operations for axis=1:
        </strong>
        <br> <!-- This adds an empty line between the two parts -->
        <h5 style="color: black;">
        * Dropping columns: If you want to remove a column (or columns), you use axis=1.<br>
        * Applying functions across rows: When performing a sum or mean across rows, you use axis=1 to apply the operation across columns for each row.

<div style="border: 2px Darkred; padding: 15px; background-color: #f9f9f9;">
    <h5 style="color: Black;">
        <strong>
            Example:
</div>

```python
df.sum(axis=1);
```

            20. In filtering, explain the following concepts:

        * Create a new column with their average:<br>
        &<br>
        |<br>

```python
# Sample DataFrame
df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60]
});

# Create a new column with the average of columns 'A' and 'B'
df['avg'] = df[['A', 'B']].mean(axis=1);

print(df);
```

```python
# Sample DataFrame
df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60]
});

# Filter rows where 'A' > 15 AND 'B' < 60
filtered_df = df[(df['A'] > 15) & (df['B'] < 60)];

print(filtered_df);
```

```python
# Filter rows where 'A' < 20 OR 'B' > 50
filtered_df = df[(df['A'] < 20) | (df['B'] > 50)]

print(filtered_df);
```

```python
# Filter rows where 'A' is not 20
filtered_df = df[~(df['A'] == 20)]

print(filtered_df);
```

```python
# Filter rows where 'A' is in the list [10, 30]
filtered_df = df[df['A'].isin([10, 30])]

print(filtered_df);
```

```python
# Filter rows where 'A' is between 10 and 25
filtered_df = df[df['A'].between(10, 25)]

print(filtered_df);
```
